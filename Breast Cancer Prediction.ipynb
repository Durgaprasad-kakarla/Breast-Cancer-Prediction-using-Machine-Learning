### Handling Outliers

```python
for feature in continuous_features:
    IQR=df[feature].quantile(0.75)-df[feature].quantile(0.25)
    print(feature,"details")
    print(IQR)
    lower_bridge=df[feature].quantile(0.25)-IQR*1.5
    print(lower_bridge)
    upper_bridge=df[feature].quantile(0.75)+IQR*1.5
    print(upper_bridge)
    df.loc[df[feature]>=upper_bridge,feature]=upper_bridge
    df.loc[df[feature]<lower_bridge,feature]=lower_bridge
```

```python
for feature in continuous_features:
    sns.boxplot(x=feature,data=df)
    plt.show()
```

```python
df.drop(['id'],axis=1,inplace=True)
```

# Feature Selection

```python
#Feature Selection with correlation
# With the following function we can select highly correlated features
#It will remove the first features that is correlated with anything other feature
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:#we are interested in absolute coefficient value
                colname=corr_matrix.columns[i]#getting the name of the column
                col_corr.add(colname)
    return col_corr
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(labels=['diagnosis'],axis=1),pd.DataFrame(df['diagnosis']),test_size=0.3,random_state=0)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
corr_features=correlation(X_train,0.8)
print(corr_features)
len(set(corr_features))
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
```

```python
X_test.drop(corr_features,axis=1,inplace=True)
```
